```python
import pandas as pd

path1 = "../data/evaluation_set/Robinson Q&A.txt"
path2 = "../data/evaluation_set/Raptor Q&A.txt"


with open(path1, 'r') as file:
    lines = file.readlines()
qna_text = ' '.join(lines)

with open(path2, 'r') as file:
    lines = file.readlines()
qna_text2 = ' '.join(lines)

print(qna_text2)
```

```python
import re
import pandas as pd

pattern = r'(Q\d+:.*?)\s+A\d+:(.*?)\s*(?=Q\d+|$)'
matches = re.findall(pattern, qna_text, re.DOTALL)

data = [{'Question': re.sub(r'^Q\d+:', '', q.strip()), 'Answer': a.strip()} for q, a in matches]

df = pd.DataFrame(data)
df
```
